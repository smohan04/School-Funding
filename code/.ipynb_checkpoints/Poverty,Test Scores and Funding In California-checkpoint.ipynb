## Background

The SAT and ACT are standardized tests that many colleges and universities in the United States require for their admissions process. This score is used along with other materials such as grade point average (GPA) and essay responses to determine whether or not a potential student will be accepted to the university.

The SAT has two sections of the test: Evidence-Based Reading and Writing and Math ([*source*](https://www.princetonreview.com/college/sat-sections)). The ACT has 4 sections: English, Mathematics, Reading, and Science, with an additional optional writing section ([*source*](https://www.act.org/content/act/en/products-and-services/the-act/scores/understanding-your-scores.html)). They have different score ranges, which you can read more about on their websites or additional outside sources (a quick Google search will help you understand the scores for each test):
* [SAT](https://collegereadiness.collegeboard.org/sat)
* [ACT](https://www.act.org/content/act/en.html)

Standardized tests have long been a controversial topic for students, administrators, and legislators. Since the 1940's, an increasing number of colleges have been using scores from sudents' performances on tests like the SAT and the ACT as a measure for college readiness and aptitude ([*source*](https://www.minotdailynews.com/news/local-news/2017/04/a-brief-history-of-the-sat-and-act/)). Supporters of these tests argue that these scores can be used as an objective measure to determine college admittance. Opponents of these tests claim that these tests are not accurate measures of students potential or ability and serve as an inequitable barrier to entry. Lately, more and more schools are opting to drop the SAT/ACT requirement for their Fall 2021 applications ([*read more about this here*](https://www.cnn.com/2020/04/14/us/coronavirus-colleges-sat-act-test-trnd/index.html)).

CollegeBoard has set benchmarks for both the Math and evidence-based reading & writing (ERW) sections of the exam ([*source*](https://collegereadiness.collegeboard.org/pdf/educator-benchmark-brief.pdf))
* Math Benchmark = 530
* ERW Benchmark = 480

1. Manually calculate mean:


```python
# Code:
test=[10,9,100,52,93,2,-9]
def list_mean(listed_values):
    summation=0
    for i in listed_values:
        summation+=i
    return summation/len(listed_values)
list_mean(test)
```

2. Manually calculate standard deviation:

    The formula for standard deviation is below:

    $$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

    Where $x_i$ represents each value in the dataset, $\mu$ represents the mean of all values in the dataset and $n$ represents the number of values in the dataset.

```python
# Code:
def list_deviation(listed_values2):
    avg=list_mean(listed_values2)
    summation_deviation=0
    for i in listed_values2:
        summation_deviation+= (i - avg)**2
    return (summation_deviation/len(listed_values2))**.5
list_deviation(test)
```

```python
# Code:
sat_2019_ca=pd.read_csv('..\\data\\sat_2019_ca.csv')
act_2019_ca=pd.read_csv('..\\data\\act_2019_ca.csv')
#https://www.cde.ca.gov/ds/fd/ec/currentexpense.asp
expenses_ca=pd.read_excel('..\\data\\ca_expenses_1819.xlsx')
#https://www.cde.ca.gov/ds/sd/sd/filessp.asp
free_eligibility=pd.read_excel('..\\data\\free1819.xlsx')
```

```python
#Expenses
expenses_ca=expenses_ca.drop(columns=['CO Code','District Code','LEA Type'])
exp_dict={
    'District':'district',
    'EDP 365':'school_year_expenditure',
    'Current\nExpense ADA':'average_daily_attendance',
    'Current\nExpense Per ADA':'daily_expenditure'
}
expenses_ca=expenses_ca.rename(columns=exp_dict)
```

```python
#SAT
#Filtering for District Data
sat_2019_ca=sat_2019_ca.loc[sat_2019_ca['RType']=='D']

#Dropping Unwanted Columns
edited_sat=sat_2019_ca.drop(columns=['SName','CName',
                                     'CDCode','CDS',
                                     'CCode','SCode',
                                     'RType','Unnamed: 25',
                                     'Year','NumERWBenchmark12',
                                     'NumMathBenchmark12','NumMathBenchmark11',
                                     'NumERWBenchmark11','NumERWBenchmark12',
                                     'PctERWBenchmark12','PctMathBenchmark12',
                                     'PctERWBenchmark11','PctMathBenchmark11',
                                     'PctBothBenchmark12','PctBothBenchmark11'])

#Creating Dictionary for Column Name Change
{name : name.lower() for name in edited_sat.columns};
sat_new_dict={
 'DName': 'district',
 'Enroll12': '12th_enrollment',
 'NumTSTTakr12': '12th_number_of_test_takers',
 'Enroll11': '11th_enrollment',
 'NumTSTTakr11': '11th_number_of_test_takers',
 'TotNumBothBenchmark12': '12th_tot_num_meeting_both_benchmark',
 'TotNumBothBenchmark11': '11th_tot_num_meeting_both_benchmark',
        }
edited_sat=edited_sat.rename(columns=sat_new_dict)
edited_sat

#Drop Null Values & *
def sat_null_fix(var):
    if var=='*':
        return np.nan
    else:
        return float(var)

edited_sat['12th_tot_num_meeting_both_benchmark']=edited_sat['12th_tot_num_meeting_both_benchmark'].map(sat_null_fix)
edited_sat['12th_number_of_test_takers']=edited_sat['12th_number_of_test_takers'].map(sat_null_fix)
edited_sat['11th_tot_num_meeting_both_benchmark']=edited_sat['11th_tot_num_meeting_both_benchmark'].map(sat_null_fix)
edited_sat['11th_number_of_test_takers']=edited_sat['11th_number_of_test_takers'].map(sat_null_fix)
edited_sat=edited_sat.dropna()

#Add New Column & Remove Unncessary Columns
temp=edited_sat['11th_tot_num_meeting_both_benchmark']+edited_sat['12th_tot_num_meeting_both_benchmark']
edited_sat['total_enrollment']=edited_sat['11th_enrollment']+edited_sat['12th_enrollment']
edited_sat['total_test_takers']=edited_sat['11th_number_of_test_takers']+edited_sat['12th_number_of_test_takers']
edited_sat['percent_meeting_both_benchmarks']=temp*100/edited_sat['total_test_takers']
edited_sat=edited_sat.drop(columns=['12th_enrollment', '12th_number_of_test_takers',
       '11th_enrollment', '11th_number_of_test_takers',
       '12th_tot_num_meeting_both_benchmark',
       '11th_tot_num_meeting_both_benchmark',])

#SAT Merge with Expenses
sat_merge_exp=pd.merge(left=edited_sat, right=expenses_ca, left_on='district', right_on='district')

#Remove Outliers SAT daily_expenditure & SAT 
sat=sat_merge_exp['daily_expenditure'].describe()
sat_iqr=sat[6]-sat[4]
sat_lower_bound=sat[4]-1.5*sat_iqr
sat_upper_bound=sat[6]+1.5*sat_iqr
sat=sat_merge_exp[sat_lower_bound<sat_merge_exp['daily_expenditure']]
sat=sat[sat['daily_expenditure']<sat_upper_bound]
```

```python
#ACT
#Filtering for District Data
act_2019_ca=act_2019_ca[act_2019_ca['RType']=='D']
edited_act=act_2019_ca.drop(columns=['CDS','CDCode', 
                                     'CCode','SCode',
                                     'RType','SName',
                                     'CName','AvgScrRead',
                                     'AvgScrEng','AvgScrMath',
                                     'AvgScrSci','Year','Unnamed: 17',
                                     'NumGE21'])
#Creating Dictionary for Column Name Change
act_new_dict={
 'DName': 'district',
 'Enroll12': '12th_enrollment',
 'NumTstTakr': '12th_number_of_test_takers',
 'PctGE21':'%_scores_greater>=21'}
edited_act=edited_act.rename(columns=act_new_dict)

#Drop Null Values & *
def act_null_fix(var):
    if var=='*':
        return np.nan
    else:
        return float(var)
edited_act['12th_enrollment']=edited_act['12th_enrollment'].map(act_null_fix)
edited_act['12th_number_of_test_takers']=edited_act['12th_number_of_test_takers'].map(act_null_fix)
edited_act['%_scores_greater>=21']=edited_act['%_scores_greater>=21'].map(act_null_fix)
edited_act=edited_act.dropna()

#ACT Merge with Expenses
act_merge_exp=pd.merge(left=edited_act, right=expenses_ca, left_on='district', right_on='district')

#Remove Outliers ACT daily_expenditure
act=act_merge_exp['daily_expenditure'].describe()
act_iqr=act[6]-act[4]
act_lower_bound=act[4]-1.5*act_iqr
act_upper_bound=act[6]+1.5*act_iqr
act=act_merge_exp[act_lower_bound<act_merge_exp['daily_expenditure']]
act=act[act['daily_expenditure']<act_upper_bound]
```

```python
#FRPM Eligibility
#Drop Unnecessary Columns & Group by District
free_elig=free_eligibility.drop(columns=['Academic Year','School Name','County Code','District Code', 'School Code', 'County Name', 'District Type','School Type','Educational \nOption Type','NSLP \nProvision \nStatus','Charter \nSchool \n(Y/N)','Charter \nSchool \nNumber','IRC','Charter \nFunding \nType','Low Grade','High Grade','Free Meal \nCount \n(K-12)','Enrollment \n(Ages 5-17)','Free Meal \nCount \n(Ages 5-17)','FRPM Count \n(Ages 5-17)','CALPADS Fall 1 \nCertification Status','Percent (%) \nEligible Free \n(K-12)','Percent (%) \nEligible Free \n(Ages 5-17)','Percent (%) \nEligible FRPM \n(Ages 5-17)','Percent (%) \nEligible FRPM \n(K-12)'])
free_elig=free_elig.groupby('District Name').sum().reset_index()
#Change Column Name & Add New Column
free_new_dict={
 'District Name': 'district',
 'Enrollment \n(K-12)': 'enrollment_k_thru_12',
 'FRPM Count \n(K-12)': 'num_students_eligible_for_frpm'}
free_elig=free_elig.rename(columns=free_new_dict)

free_elig['percentage_eligible_for_frpm']=free_elig['num_students_eligible_for_frpm']/free_elig['enrollment_k_thru_12']*100
```

```python
#Merging with FRPM Eligibility
act_frpm=pd.merge(left=act, right=free_elig, left_on='district', right_on='district')
sat_frpm=pd.merge(left=sat, right=free_elig, left_on='district', right_on='district')
#Getting Rid of Duplicates
act_frpm[act_frpm.duplicated('district')]
sat_frpm[sat_frpm.duplicated('district')]
act_frpm=act_frpm.drop_duplicates(subset='district',keep=False)
sat_frpm=sat_frpm.drop_duplicates(subset='district',keep=False)
```

```python
#Saving As CSV
sat_frpm.to_csv('..//data//SAT_FRPM_EXP.csv') 
act_frpm.to_csv('..//data//ACT_FRPM_EXP.csv')
```
